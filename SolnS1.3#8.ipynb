I added some color to the picture but the real issue was just how fast $e^{xy}$ blows up.  I was using 10 on the limits for $x$ and $y$ so it intepreted every other small number as zero.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-1,1,11),np.linspace(-1,1,11))

u = 1
v = np.round(np.exp(x*y),10)

gradmag = np.sqrt(u**2+v**2)

plt.pcolor(x, y, gradmag,cmap='rainbow')
plt.colorbar()

plt.quiver(x,y,u,v)
plt.show()
```
