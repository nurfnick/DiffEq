# Vector Plots for Differential Equations

Here is the basic plotting ideas for a vectorfield.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = 1
v = -1

plt.quiver(x,y,u,v)
plt.show()
```

Let's get a differential equation involved.

$$
y' = ay
$$

Then the slope of the field at any point is $m =\frac{\Delta y}{\Delta x}= y$ so we'll set the $u$ from above ($x$-component) equal to the denominator and the $v$ equal to the numerator.  This should give us the vector field corresponding to this diff eq

```python
a = 1 # I needed to set a equal to something...

u = 1
v = a*y

plt.quiver(x,y,u,v)
plt.show()
```

If we know the solution $y = ce^{ax}$.  We can add that to the flow.  Again I had to pick a value for $c$.

```python
x1 = np.linspace(0,2,1000)

u = 1
v = a*y

plt.quiver(x,y,u,v)
plt.plot(x1,np.exp(x1),'r')
plt.show()
```

##Newton's Law of Cooling

$$ T' = -k(T-T_m)$$

```python
Tm = 0
k = 1

u = np.ones(y.shape)
v = -k*(y-Tm)

plt.quiver(x,y,u,v)
plt.show()
```

```python
Tm = 0
k = 1

u = np.ones(y.shape)
v = -k*(y-Tm)

plt.streamplot(x,y,u,v,arrowstyle='-')
plt.quiver(x,y,u,v)
plt.show()
```
