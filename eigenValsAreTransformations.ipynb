```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 2*np.pi +0.1, 0.1)

x,y = np.sin(t), np.cos(t)

plt.plot(x,y)
plt.show()

A = np.array([[4,-3],[2,-3]])

def transform(x,y):
    return A[0,0]*x + A[0,1]*y, A[1,0]*x + A[1,1]*y

x,y = transform(x,y)

plt.plot(x,y)
plt.show()
```

```python
eigens = np.linalg.eig(A)
print(eigens)
```
